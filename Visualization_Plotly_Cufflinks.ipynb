```python
df = pd.DataFrame(np.random.randn(100,3), columns = ['A', 'B', 'C'])
df.head()
df['A'] = df['A'].cumsum() + 20
df['B'] = df['B'].cumsum() + 20
df['C'] = df['C'].cumsum() + 20
```

```python
df.head()
```

```python
df.plot()  #pandas Plot
```

```python
#to initialize offline notebook mode-- reason- cufflink is configuration library for potly..cufflink provides both online and offline 
#configuration as well...here we are using local system thats why using offline mode.
#when you will click on export to potly given at bottom of graph by potly...it will take you online potly..

py.offline.init_notebook_mode(connected=True)
cf.go_offline()
```

```python
#potly iplot
df.iplot() 
#move curson on pgraph it is intereactive and click on export to potly
```

```python
df.iplot(x='A', y='B', mode='markers', size=15)
```
