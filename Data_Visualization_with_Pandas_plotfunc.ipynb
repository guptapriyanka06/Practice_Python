# scatter matrix with kde useful for Normal Distribution

# kde=> diagonal grapgs(same col vs same col) showing data for a each col is normaly distributed or not?
if dataset is not normalized then for sure you have to normalize the data for that particular feature.
in below scatter matrix, only one col is normalized which is bell shaped curve.

```python
pd.plotting.scatter_matrix(iris,figsize=(10,10),color='r',alpha=.9,diagonal='kde')
```
